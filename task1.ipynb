```python
# 1.Load the dataset using Pandas and display first and last few records to understand the structure of rows and columns.
import pandas as pd
import numpy as np
df1=pd.read_csv("titanic.csv")
df1.head()
df1.tail()

df2=pd.read_csv("student_performance.csv")
df2.head()
df2.tail()
```

```python
# 2.Identify numerical, categorical, ordinal, and binary features manually by inspecting column names and values.
print(df1.dtypes)
print(df1.select_dtypes(include=np.number).columns)
print(df1.select_dtypes(exclude=np.number).columns)
print(df1['survived'].unique())
print(df1['pclass'].unique())

print(df2.dtypes)
print(df2.select_dtypes(include=np.number).columns)
print(df2.select_dtypes(exclude=np.number).columns)
print(df2['grade'].unique())
```

```python
# 3.Use df.info() and df.describe() to understand data types, null values, and statistical summaries.
print(df1.info())
print(df1.describe())

print(df2.info())
print(df2.describe())
```

```python
# 4.Check unique values in categorical columns to understand data distribution.
print(df1['sex'].unique())
print(df1['embarked'].unique())
print(df1['pclass'].unique())

print(df2['grade'].unique())
```

```python
# 5.Identify target variable and input features for ML suitability.
target_titanic='survived'
features_titanic = df1.drop(columns=['survived'])
print(target_titanic)
print(features_titanic.columns)

target_student='grade'
features_student=df2.drop(columns=['grade'])
print(target_student)
print(features_student.columns)
```

```python
# 6.Analyze dataset size and discuss whether it is suitable for machine learning.
print(df1.size)
print(len(df1))

print(df2.size)
print(len(df2))
```

```python
# 7.Write clear observations about data quality issues like missing values or imbalance.
print(df1.isnull().sum())
print(df1['survived'].value_counts())
print(df1['survived'].value_counts(normalize=True)*100)

print(df2.isnull().sum())
print(df2['grade'].value_counts())
print(df2['grade'].value_counts(normalize=True)*100)
```
